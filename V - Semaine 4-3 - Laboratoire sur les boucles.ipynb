# Question 1
Votre programme doit demander une quantité inconnu de nombre et afficher:
- Le nombre le plus grand
- Le nombre le plus petit
Votre programme doit quitte lorsque le nombre entré est zéro.

# Question 2
Il s'agit de faire un jeu à 2 joueurs pour deviner un nombre. Le premier joueur devra entrer le nombre à faire deviner et le second tentera de trouver le nombre. Pour aider, le programme devra indiquer s'il faut monter ou descendre. Le pointage final équivaut au nombre de tentatives.

## Suggestion
Pour faire "disparaitre" le nombre choisi par le premier joueur, effectuer 100 print vide après avoir obtenu le nombre.


# Question 3
Vous devez effectuer un programme de remboursement de paiement.  Un montant initial doit être demandé à des fins de dette. Par la suite, le programme demandera les montants à rembourser jusqu'à ce que la totalité soit remboursée. Des statistiques s'afficheront à la fin pour indiquer le nombre de paiement effectués ainsi que la moyenne de chaque paiement.

## Note de validation
Si le dernier paiement est supérieur à la dette restante, seul la dette est remboursée. On ne doit pas rembourser plus que la dette.

## Résultat attendu
```Quel est le montant de la dette? ``` *1000*  
`********************************`  
```Dette: 1000$. Combien voulez-vous rembourser?```  *50*  
```Dette:  950$. Combien voulez-vous rembourser?``` *900*  
```Dette:   50$. Combien voulez-vous rembourser?``` *100*  
```Dernier remboursement de 50$ effectué.```  
`*************`    
```
Nombre de paiements:            3
Montant total remboursé: 1000.00$
Moyenne des paiements:    333.33$
```
